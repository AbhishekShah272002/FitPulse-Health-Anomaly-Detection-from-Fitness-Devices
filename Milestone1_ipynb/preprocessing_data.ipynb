ML Preprocessing Steps

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce", dayfirst=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
df.numeric_columns
```

```python
df.info()
```

```python
df.groupby("User_ID")[["Hours_Slept", "Steps_Taken", "Water_Intake (Liters)", "Active_Minutes", "Heart_Rate (bpm)"]].apply(lambda x: x.isnull().sum())
```

```python
numeric_cols = ["Hours_Slept", "Steps_Taken", "Water_Intake (Liters)", "Active_Minutes", "Heart_Rate (bpm)"]

df[numeric_cols] = df.groupby("User_ID")[numeric_cols].transform(lambda x: x.interpolate(method="linear"))
```
